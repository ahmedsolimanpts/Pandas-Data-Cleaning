```python
import pandas as pd
import numpy as np

class DataCleaning:
    
    def ReadCsvFile(self,filename): # Function To Read CSV File
        try:
            df = pd.read_csv(filename)
            return df
        except  :
            return 'Sorry Error Happen check the File Name'
        
    # ---------------------------------------------------------------------------------------------------- #
   
    def ConnecttoMysql(self,user,password,port,database): # Function To Create Engine For MYSQL
        from sqlalchemy import  create_engine
        try:
            link ='mysql+pymysql://{}:{}@localhost:{}/{}'.format(user,password,port,database)
            engine = create_engine(link)
            return engine
        except:
             'Sorry Error Happen check the Inputs'
            
    
    # ---------------------------------------------------------------------------------------------------- #
    
    def ReadFromMySQL(self,table,engine): # Function To Get Table From MYSQL
        try:
            df = pd.read_sql_query('Select * from {}'.format(table),engine)
            return df
        except Exception as e:
            return e
    
    
    # ---------------------------------------------------------------------------------------------------- #
    
    
    def ReadEcelFile(self,filename): # Function To Read Excel File
         try:
            df = pd.read_excel(filename)
            return df
         except :
            return 'Sorry Error Happen check the File Name'
        
    # ---------------------------------------------------------------------------------------------------- #

    def DropColumn(self,df,column): # Function To Drop Column
        try:
            df =df.drop(column,axis=1)
            return df
        except:
            return 'Sorry Error Happen check the Column Name or Data Frame Name'

    # ---------------------------------------------------------------------------------------------------- #
    
    def GetSumofNan(self,df): # Function To Get All Missing Value IN Columns
        nans= df.isnull().sum()
        return nans

    # ---------------------------------------------------------------------------------------------------- #
    
    def ReplaceinCol(self,df,column,dictinary): # Function to Replace Values in One Column 
        try:
            df[column].replace(dictinary,inplace=True)
            return df
        except:
            return 'Sorry Error Happen check the Column Name'
    
    # ---------------------------------------------------------------------------------------------------- #
    
    def GetUniqueValueinColumn(self,df,column): # Function To Get The ALL Unique Value In One Column
        try:
            return df[column].unique()
        except:
            return 'Sorry Error Happen check the Column Name'
        
    # ---------------------------------------------------------------------------------------------------- #
    
    def RenameColumn(self,df,oldname,newname): # Function To Rename Column
        try:
             df.rename(columns={oldname:newname},inplace=True)
        except:
            return 'Sorry Error Happen check the Columns Names Or Data Frame Name'
        
    # ---------------------------------------------------------------------------------------------------- #
    
    def DropDuplicate(self,df): # Function To Drop Duplicates Row IN DataFrame
        try:
            df.drop_duplicates(inplace=True)
        except:
            return 'Sorry Error Happen check the Data Frame Name'
        
    # ---------------------------------------------------------------------------------------------------- #
    
    def DropNan(self,df,how): # Function To Drop NAN Value Either All in Row Have NAN or Only One in Row
        try:
            df =df.dropna(how=how)
            return df
        except Exception as e:
            return e
    
```
