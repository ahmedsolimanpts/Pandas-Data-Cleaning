```python
import pandas as pd
import numpy as np

class DataCleaning:
    
    # Function To Read CSV File
    def ReadCsvFile(self,filename): # Parameters is File Name
        try:
            dataframe = pd.read_csv(filename)
            return dataframe
        except Exception as e:
            return e  
    
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Create Engine For MYSQL
    def ConnecttoMysql(self,user,password,port,database_name):  # Parameters is user Name,Password,Port AND DATABASE NAME
        from sqlalchemy import  create_engine
        try:
            link ='mysql+pymysql://{}:{}@localhost:{}/{}'.format(user,password,port,database_name)
            engine = create_engine(link)
            return engine
        except Exception as e:
            return e
   
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Get Table From MYSQL
    def ReadFromMySQL(self,table,engine): # Parameters is Table Name and Engine Name That Create Before
        try:
            dataframe = pd.read_sql_query('Select * from {}'.format(table),engine)
            return dataframe
        except Exception as e:
            return e
    
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Read Excel File
    def ReadEcelFile(self,file_name):  # Parameters is File Name
         try:
            df = pd.read_excel(file_name)
            return df
         except Exception as e:
            return e
   
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Drop Column
    def DropColumn(self,dataframe,column_name): # Parameters is DataFrame and Columns Name
        try:
            dataframe =dataframe.drop(column_name,axis=1)
            return dataframe
        except Exception as e:
            return e
    
    # ---------------------------------------------------------------------------------------------------- #    
    
    # Function To Get  Sum of All Missing Value IN Columns
    def GetSumofNan(self,DataFrame): # Parameters is DataFrame 
        try:
            nans= DataFrame.isnull().sum()
            return nans
        except Exception as e:
            return e
        
    # ---------------------------------------------------------------------------------------------------- #  
    
    # Function to Replace Values in One Column 
    def ReplaceinCol(self,dataframe,column_name,dictinary): # Parameters is DataFrame , Column Name and Dict Of old Values and New Values
        try:
            dataframe[column_name].replace(dictinary,inplace=True)
            return dataframe
        except Exception as e:
            return e
    
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Get The ALL Unique Value In One Column
    def GetUniqueValueinColumn(self,dataframe,column_name): # Parameters is DataFrame and Column Name
        try:
            return dataframe[column_name].unique()
        except Exception as e:
            return e
    
    # ---------------------------------------------------------------------------------------------------- #
   
    # Function To Rename Column
    def RenameOneColumn(self,dataframe,old_name,new_name):  # Parameters is DataFrame , Old Column Name and New Column Name
        try:
             dataframe.rename(columns={old_name:new_name},inplace=True)
        except Exception as e:
            return e        
    
    # ---------------------------------------------------------------------------------------------------- #
   
    # Function To Drop Duplicates Row IN DataFrame
    def DropDuplicate(self,dataframe): # Parameters is DataFrame 
        try:
            dataframe.drop_duplicates(inplace=True)
        except Exception as e:
            return e 
    
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Drop NAN Value Either All in Row Have NAN or Only One in Row
    def DropNan(self,dataframe,how): # Parameters is DataFrame , How we Drop Nan (Any,ALL) 
        try:
            dataframe = dataframe.dropna(how=how)
            return dataframe
        except Exception as e:
            return e
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Concat two Data Frame 0 For Add on Tail 1 For Add in New Columns
    def concat2table(self,first_dataframe,second_dataframe,how): # Parameters is First DataFrame ,Second DataFrame and How we Drop Nan (Any,ALL) 
        try:    
            data =pd.concat([first_dataframe,second_dataframe],axis=how)
            return data
        except Exception as e:
            return e
   
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function To Join Two Table ( Get Intersiction )
    def intersectiontable(self,first_dataframe,second_dataframe): # Parameters is First DataFrame ,Second DataFrame 
        try:    
            new_data =pd.merge(first_dataframe,second_dataframe)
            return new_data
        except Exception as e:
            return e
 
    # ---------------------------------------------------------------------------------------------------- #
   
    # Function to Rename Columns in the DataFrame
    def RenameColumns(self,DataFrame,Dictinary):  # Parameters is the DataFrame and Dictinary of Old_Name and New_Name                                 
        try:
            DataFrame.rename( columns = Dictinary ,inplace = True )
        except Exception as e:
            return e
    
    # ---------------------------------------------------------------------------------------------------- #
    
    # Function to Count Vaues In Column
    def CountColumnValues(self,dataframe,column_name): # Parameters is Data Frame and Columns Name
        try:
            return dataframe[column_name].value_counts()
        except Exception as e:
            return e
            
```
